```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x = np.arange(100)
f1 = lambda x:x**2
plt.figure()
plt.plot(x,f1(x))
plt.show()
```

```python
f2 = lambda x:np.sin(x)
x = np.arange(-3,3,0.01)
plt.figure()
plt.plot(x,f2(x))
plt.show()
```

```python
f3 = lambda x:np.sinc(x)
x = np.arange(-3,3,0.01)
plt.figure()
plt.plot(x,f3(x))
plt.show()
```
